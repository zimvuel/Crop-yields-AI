```python
import ee
import pandas as pd

# 1. Inisialisasi GEE
try:
    ee.Initialize()
except Exception as e:
    ee.Authenticate()
    ee.Initialize()

print("‚úÖ Terhubung ke Google Earth Engine.")

# 2. Data Koordinat (Lengkap)
provinsi_coords = {
    "Aceh": [96.7207, 4.3685],
    "Sumatera Utara": [99.1320, 2.3822],
    "Sumatera Barat": [100.6468, -0.9057],
    "Riau": [101.5974, 0.4773],
    "Jambi": [103.1043, -1.6252],
    "Sumatera Selatan": [104.1389, -3.2839],
    "Bengkulu": [102.4478, -3.6380],
    "Lampung": [105.1365, -4.8461],
    "Kep. Bangka Belitung": [106.4453, -2.4603],
    "Kep. Riau": [106.2964, 2.2085], # Target Utama Perbaikan
    "DKI Jakarta": [106.8456, -6.2088], # Target Utama Perbaikan (Urban)
    "Jawa Barat": [107.6641, -6.9594],
    "Jawa Tengah": [110.1403, -7.1510],
    "DI Yogyakarta": [110.4262, -7.8754],
    "Jawa Timur": [112.1753, -7.7544],
    "Banten": [106.1093, -6.4058],
    "Bali": [115.1317, -8.3698],
    "NTB": [117.3954, -8.6349],
    "NTT": [121.7171, -8.5571],
    "Kalimantan Barat": [111.4667, -0.0894],
    "Kalimantan Tengah": [113.5693, -1.5382],
    "Kalimantan Selatan": [115.3798, -2.8552],
    "Kalimantan Timur": [116.5502, 0.8478],
    "Kalimantan Utara": [116.3478, 3.0453],
    "Sulawesi Utara": [124.4160, 0.9681],
    "Sulawesi Tengah": [120.8888, -1.1852], # Target Utama Perbaikan
    "Sulawesi Selatan": [120.0220, -3.7724],
    "Sulawesi Tenggara": [122.0974, -4.0467],
    "Gorontalo": [122.3662, 0.7066],
    "Sulawesi Barat": [119.3487, -2.4878],
    "Maluku": [129.3900, -3.3446],
    "Maluku Utara": [127.9739, 0.8273],
    "Papua": [138.0723, -2.5028],
    "Papua Barat": [133.2774, -1.4639],
    "Papua Selatan": [139.5670, -7.1370],
    "Papua Tengah": [136.2821, -3.6203],
    "Papua Pegunungan": [139.3923, -4.2453],
    "Papua Barat Daya": [131.5831, -1.2567]
}

# 3. Definisi Dataset ISRIC SoilGrids
try:
    img_ph = ee.Image("projects/soilgrids-isric/phh2o_mean").select('phh2o_0-5cm_mean')
    img_clay = ee.Image("projects/soilgrids-isric/clay_mean").select('clay_0-5cm_mean')
    img_sand = ee.Image("projects/soilgrids-isric/sand_mean").select('sand_0-5cm_mean')
    img_soc = ee.Image("projects/soilgrids-isric/soc_mean").select('soc_0-5cm_mean')
except Exception as e:
    print("Error dataset GEE.")

def get_soil_data_adaptive(provinsi, lon, lat):
    # --- STRATEGI BARU: ADAPTIVE LOOPING ---
    # Kita coba radius bertahap: 10km -> 30km -> 60km -> 100km
    # Ini menjamin provinsi kecil dapat data presisi, provinsi sulit dapat data luas
    buffer_sizes = [10000, 30000, 60000, 100000] 
    
    final_data = None
    used_radius = 0
    
    for radius in buffer_sizes:
        try:
            point = ee.Geometry.Point([lon, lat]).buffer(radius)
            
            # Hitung statistik
            # bestEffort=True sangat penting agar tidak error saat radius membesar
            ph_val = img_ph.reduceRegion(reducer=ee.Reducer.mean(), geometry=point, scale=1000, bestEffort=True).get('phh2o_0-5cm_mean')
            clay_val = img_clay.reduceRegion(reducer=ee.Reducer.mean(), geometry=point, scale=1000, bestEffort=True).get('clay_0-5cm_mean')
            sand_val = img_sand.reduceRegion(reducer=ee.Reducer.mean(), geometry=point, scale=1000, bestEffort=True).get('sand_0-5cm_mean')
            soc_val = img_soc.reduceRegion(reducer=ee.Reducer.mean(), geometry=point, scale=1000, bestEffort=True).get('soc_0-5cm_mean')
            
            # Ambil nilai
            p = ph_val.getInfo()
            
            # Jika p ada nilainya (bukan None), berarti kita berhasil dpt tanah!
            if p is not None:
                c = clay_val.getInfo()
                s = sand_val.getInfo()
                o = soc_val.getInfo()
                
                final_data = {
                    "Provinsi": provinsi,
                    "Lat": lat,
                    "Lon": lon,
                    "pH": p / 10.0,
                    "Clay_%": c / 10.0 if c else None,
                    "Sand_%": s / 10.0 if s else None,
                    "Organic_Carbon_g/kg": o / 10.0 if o else None,
                    "Radius_Used_km": radius / 1000 # Info tambahan: radius yang dipakai
                }
                used_radius = radius
                break # Keluar dari loop karena sudah dapat data
                
        except Exception:
            continue # Coba radius berikutnya jika error

    if final_data is None:
        print(f"‚ùå Gagal total untuk {provinsi} bahkan dengan radius 100km.")
    elif used_radius > 10000:
        # Info jika harus memperluas pencarian (untuk Jakarta/Kep Riau biasanya muncul ini)
        print(f"   ‚ö† {provinsi}: Sukses dengan perluasan radius ke {used_radius/1000} km")
        
    return final_data

# 4. Eksekusi
results = []
print("\nüöÄ Memulai pengambilan data dengan Adaptive Buffer...")

for prov, coords in provinsi_coords.items():
    # Feedback visual sederhana
    print(f"   Processing: {prov}...", end="\r")
    data = get_soil_data_adaptive(prov, coords[0], coords[1])
    if data:
        results.append(data)

print("\n\n‚úÖ Selesai memproses semua provinsi.")

# 5. Simpan
if results:
    df_final = pd.DataFrame(results)
    print("\nPreview Hasil:")
    print(df_final[['Provinsi', 'pH', 'Clay_%', 'Radius_Used_km']].head())
    
    df_final.to_csv("Data_Tanah_Indonesia_Adaptive_Fixed.csv", index=False)
    print("\nüìÇ File tersimpan: 'Data_Tanah_Indonesia_Adaptive_Fixed.csv'")
else:
    print("Tidak ada data.")
```

```python
import ee
import pandas as pd
from datetime import datetime, timedelta
from dateutil.relativedelta import relativedelta

# 1. Inisialisasi GEE
try:
    ee.Initialize()
except Exception as e:
    ee.Authenticate()
    ee.Initialize()

print("‚úÖ Terhubung ke Google Earth Engine.")

# 2. Konfigurasi Provinsi (Radius Besar)
provinsi_config = {
    "Aceh": [96.7207, 4.3685, 50000],
    "Sumatera Utara": [99.1320, 2.3822, 50000],
    "Sumatera Barat": [100.6468, -0.9057, 50000],
    "Riau": [101.5974, 0.4773, 50000],
    "Jambi": [103.1043, -1.6252, 50000],
    "Sumatera Selatan": [104.1389, -3.2839, 50000],
    "Bengkulu": [102.4478, -3.6380, 50000],
    "Lampung": [105.1365, -4.8461, 50000],
    "Kep. Bangka Belitung": [106.4453, -2.4603, 60000],
    "Kepulauan Riau": [106.2964, 2.2085, 100000],
    "DKI Jakarta": [106.8456, -6.2088, 30000],
    "Jawa Barat": [107.6641, -6.9594, 30000],
    "Jawa Tengah": [110.1403, -7.1510, 30000],
    "DI Yogyakarta": [110.4262, -7.8754, 30000],
    "Jawa Timur": [112.1753, -7.7544, 30000],
    "Banten": [106.1093, -6.4058, 30000],
    "Bali": [115.1317, -8.3698, 30000],
    "NTB": [117.3954, -8.6349, 50000],
    "NTT": [121.7171, -8.5571, 50000],
    "Kalimantan Barat": [111.4667, -0.0894, 50000],
    "Kalimantan Tengah": [113.5693, -1.5382, 50000],
    "Kalimantan Selatan": [115.3798, -2.8552, 50000],
    "Kalimantan Timur": [116.5502, 0.8478, 50000],
    "Kalimantan Utara": [116.3478, 3.0453, 50000],
    "Sulawesi Utara": [124.4160, 0.9681, 50000],
    "Sulawesi Tengah": [120.8888, -1.1852, 30000],
    "Sulawesi Selatan": [120.0220, -3.7724, 50000],
    "Sulawesi Tenggara": [122.0974, -4.0467, 50000],
    "Gorontalo": [122.3662, 0.7066, 50000],
    "Sulawesi Barat": [119.3487, -2.4878, 50000],
    "Maluku": [129.3900, -3.3446, 60000],
    "Maluku Utara": [127.9739, 0.8273, 60000],
    "Papua": [138.0723, -2.5028, 50000],
    "Papua Barat": [133.2774, -1.4639, 50000],
    "Papua Selatan": [139.5670, -7.1370, 50000],
    "Papua Tengah": [136.2821, -3.6203, 50000],
    "Papua Pegunungan": [139.3923, -4.2453, 50000],
    "Papua Barat Daya": [131.5831, -1.2567, 50000]
}

# 3. Dataset & Fungsi Core
DATASET_ID = "NASA/SMAP/SPL4SMGP/008"
BAND_NAME = "sm_surface"

def get_data_chunk(provinsi, lon, lat, radius, start_date, end_date):
    """Mengambil data untuk rentang waktu pendek (Chunk)"""
    area = ee.Geometry.Point([lon, lat]).buffer(radius)
    
    smap = ee.ImageCollection(DATASET_ID)\
             .select(BAND_NAME)\
             .filterDate(start_date, end_date)

    def extract_val(img):
        date = img.date().format('YYYY-MM-dd')
        stats = img.reduceRegion(
            reducer=ee.Reducer.mean(),
            geometry=area,
            scale=11000, 
            bestEffort=True
        )
        return ee.Feature(None, {
            'Date': date,
            'Moisture_Raw': stats.get(BAND_NAME),
            'Provinsi': provinsi
        })

    try:
        # getInfo() di sini aman karena hanya menarik data 3 bulan
        return smap.map(extract_val).getInfo()
    except Exception as e:
        return None

# 4. EKSEKUSI BATCHING (Pemecahan Waktu)
# Kita pecah per 3 bulan agar memori GEE tidak jebol
start_date = datetime(2023, 11, 1)
end_date = datetime(2025, 12, 31)

# Buat list rentang waktu per 3 bulan
chunks = []
curr = start_date
while curr < end_date:
    next_curr = curr + relativedelta(months=3) # Batch 3 Bulan
    if next_curr > end_date:
        next_curr = end_date
    chunks.append((curr.strftime("%Y-%m-%d"), next_curr.strftime("%Y-%m-%d")))
    curr = next_curr

all_results = []

print(f"üöÄ Memulai Batch Processing (Total {len(chunks)} batch waktu per provinsi)...")

for prov, params in provinsi_config.items():
    lon, lat, rad = params
    print(f"\nüìç {prov} (Radius {rad/1000} km)")
    
    prov_data = []
    for i, (s_date, e_date) in enumerate(chunks):
        # Print progress bar sederhana
        print(f"   - Batch {i+1}/{len(chunks)}: {s_date} s/d {e_date} ...", end="\r")
        
        chunk_res = get_data_chunk(prov, lon, lat, rad, s_date, e_date)
        
        if chunk_res and 'features' in chunk_res:
            for feat in chunk_res['features']:
                p = feat['properties']
                if p.get('Moisture_Raw') is not None:
                    prov_data.append(p)
    
    if prov_data:
        print(f"   ‚úÖ Selesai. Dapat {len(prov_data)} titik data.")
        all_results.extend(prov_data)
    else:
        print(f"   ‚ö†Ô∏è Gagal/Kosong untuk {prov}.")

# 5. Simpan & Rapikan
if all_results:
    print("\n‚öôÔ∏è Memproses data akhir...")
    df = pd.DataFrame(all_results)
    
    # Rata-rata Harian (karena data asli per 3 jam)
    df_daily = df.groupby(['Provinsi', 'Date'])['Moisture_Raw'].mean().reset_index()
    
    # Konversi ke Persen
    df_daily['Soil_Moisture_%'] = df_daily['Moisture_Raw'] * 100
    
    # Sort
    df_daily['Date'] = pd.to_datetime(df_daily['Date'])
    df_daily = df_daily.sort_values(by=['Provinsi', 'Date'])
    
    filename = "Data_Soil_Moisture_Harian_2023-2025_BATCHED.csv"
    df_daily.to_csv(filename, index=False)
    print(f"\nüéâ SUKSES BESAR! File tersimpan: '{filename}'")
    print(df_daily.head())
else:
    print("‚ùå Tetap gagal. Kemungkinan server GEE sedang down atau kuota habis.")
```
